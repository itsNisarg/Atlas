```python
import networkx as nx
import node2vec as n2v
```

```python
graph = nx.read_graphml("../networks/graphml/countries_network.graphml")
```

```python
for s, t in graph.edges():
    if s == "Azerbaijan": print(t)
```

```python
def fun(p, q):
    # Genereate Walks
    node2vec = n2v.Node2Vec(graph, dimensions=64, walk_length=50, num_walks=500, p=p, q=q)
    # Embed nodes
    model = node2vec.fit(window=10, min_count=1, batch_words=4)
    # Save embeddings for later use
    model.wv.save(f"node/node2vec_64_50_500_p{p}_q{str(q).replace('.', '_')}.kv")
    # Save model for later use
    model.save(f"model/node2vec_64_50_500_p{p}_q{str(q).replace('.' , '_')}.model")
    # Embed edges
    edge_embeds = n2v.edges.HadamardEmbedder(model.wv)
    edges_kv = edge_embeds.as_keyed_vectors()
    # Save embeddings for later use
    edges_kv.save(f"edge/edge_embeddings_64_50_500_p{p}_q{str(q).replace('.', '_')}.kv")
    
```

```python
fun(1, 0.1)
```
